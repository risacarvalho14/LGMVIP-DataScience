```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
img = cv2.imread('Flowers.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(img_gray, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(10,10))
plt.imshow(img_invert, cmap = "gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
img_smoothing = cv2.GaussianBlur(img_invert,(25,25),sigmaX = 0, sigmaY = 0)
plt.figure(figsize=(10,10))
plt.imshow(img_smoothing, cmap = "gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
final = cv2.divide(img_gray,255-img_smoothing, scale = 255)
plt.figure(figsize=(10,10))
plt.imshow(final , cmap = 'gray')
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

```python
plt.figure(figsize= (10,10))
plt.subplot(2,5,1)
plt.imshow(img)
plt.axis("off")

plt.title("Original Image")
plt.subplot(2,5,2)
plt.imshow(img_gray, cmap = "gray")
plt.axis("off")

plt.title("Gray Scale Image")
plt.subplot(2,5,3)
plt.imshow(img_invert, cmap = "gray")
plt.axis("off")

plt.title("Inverted Image")
plt.subplot(2,5,4)
plt.imshow(img_smoothing, cmap = "gray")
plt.axis("off")

plt.title("Smoothen image")
plt.subplot(2,5,5)
plt.imshow(final, cmap = "gray")
plt.axis("off")

plt.title("Final sketch Image")
plt.show()
```
